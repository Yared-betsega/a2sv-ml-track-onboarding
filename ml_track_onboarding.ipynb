# Pandas

**First Import the necessary packages**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**1. Parse the .csv file into a data frame**

```python
# Read the data and print the first few elements.

df = pd.read_csv("imdb_top_1000.csv")
df.head()
```

**Answer the following questions**

***A: What are the column names in the dataset?***

```python
# The df object have a property called columns that returns the columns in the dataset, 
# I can use that to print the columns

columns = df.columns
columns
```

***B: What is the average rating of the movies?***

```python
# Pandas by default implements some mathematical operations for each columns in the dataset. 
# We can use the mean methode from the pandas object for any column in the dataset.

ave_rating = df['IMDB_Rating'].mean()
ave_rating
```

***C: How many unique genres are there?***

```python
# use the nunique methode to count the number of unique values in a column

unique_genres = df['Genre'].nunique()
unique_genres
```

**Save your results as a csv file.**

```python
# First create a dataframe using pandas library, it will contain the results we listed above.
# convert the dataframe in to csv file using the .to_csv() methode of pandas dataframe object.

results = pd.DataFrame({
    "Column_names": columns,
    "average_rating": ave_rating,
    "unique_genres":unique_genres,
})

results
```

```python
results.to_csv('results.csv', index=False)
```

# Plotting

**Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.**

```python
# 
genre = df['Genre']

genre_counts = genre.value_counts()
sample_genres = genre_counts.head(20)

colors = ['brown']

plt.bar(sample_genres.index, sample_genres.values, color=colors)
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

**Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)**

```python
# Get the movie count for each year and sort it using the year as a key.
year_counts = df['Released_Year'].value_counts().sort_index()

# Plot the graph
colors = ['brown']
plt.figure(figsize=(20, 6))
plt.scatter(year_counts.index, year_counts.values, color=colors)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies vs Year')
plt.xticks(rotation=90,)
plt.yticks(rotation=0, )


plt.show()
```
